# 总结

这本书从总体上是这样写的。
- 过程抽象
   - 程序设计基本元素
      - 基本表达形式
      - 组合机制
      - 抽象机制
   - 高阶函数做抽象，过程和数据没有明确的区分，
      - 过程作为参数
      - 过程作为一般性的方法
      - 过程作为返回值
   - lambda ： 可以做匿名函数。
   - 时间复杂度和空间复杂度 ： 递归尽量为线性递归或者迭代。
      - 递归迭代区别
         - 递归是一个事实，即函数中会调用自己。
         - 迭代，重复反馈过程的活动，每一次迭代的结果会作为下一次迭代的初始值，也就是不断的更新几个状态变量。
- 数据抽象
   - 基本思想
      - 按愿望思维，先假设有这个复合数据对象的构造函数和选择函数，然后设法构造出一些使用复合数据对象的程序，使它们可以像操作简单数据对象类型一样操作“抽象数据”
      - 复合数据对象：
         - 构造函数 : 
         - 选择函数 ： 选择这个复合数据对象中某个或者某些元素的值。
   - 数据抽象方式
      - 数据导向 ： 函数式编程，固定操作名称，通过参数类型列表匹配函数。
      - 消息传递 ： 面向对象编程，基于给定的对象，通过操作名称匹配函数。
- 模块化/对象/状态
   - 赋值和局部状态 ： 
      - 函数式编程中，同样的参数，调用同一个函数，结果会只有一个。
      - 引进赋值后，同样的参数，调用同一个函数，结果会不止一个，这个是模拟现实世界的代价。
   - 求值的环境模型 ： 引用这个是因为赋值后会修改变量。
   - 用变动的数据做模拟。
      - 简单点就是一个对象中有相关对象的引用，如果这个对象有变动，就会通知相关的对象。
   - 并发问题 ： 实质上就是时间问题。
   - 流 ： 
      - 作为延时的表
      - 简单点就是发送部分数据给下一层，其他数据仅仅是约定以后会求值。
         - delay : 延时计算，约定以后会计算
            - 思想 ： (delay \<exp>) 可以看作 (lambda () \<exp>) ，delay直接用define是不可以实现的，需要用到宏的方式来定义,因为scheme默认不是惰性求值.
         - force ：结束这种延时。
            - 思想 ： (define (force delayed-object) (delayed-object)) ，这样会强迫求值了。
- 元语言抽象，其实就是设计一门新语言。
   - 元语言的求值器
      - 最基本的就是eval和apply循环
         - eval ： 在一个环境里求值一个表达式
         - apply ： 将一个过程对象应用于一组实际参数。
      - 分析和执行分离
         - 思想 ： 分析过程的返回是一个labmda(env)表达式，执行的之后执行这恶lambda表达式
   - 惰性求值
      - 应用序求值（先求值参数，然后应用），
      - 正则序求值，惰性求值，先展开过程再求值。
   - 非确定性计算
      - 我感觉就是暴力搜索，只是在最上层的表达式会非常的抽象，通俗易懂。
         - 我们需要从一个表取出一个数字，在另一个表取出一个数字，然后两个数字相加等于素数    
         ```scheme  
         (define (prime-sum-pair list1 list2)   
             (let ((a (an-element-of list1))   
                  (b (an-element-of list2)))  
             (require (prime? (+ a b)))  
             (list a b)))```
      - 思想：
         - 允许一个表达式里有多个可能的值。
         - 从中自动选择一个值，
         - 如果随后的要求无法满足，求值器就会回到有关的表里再次选择，直至求值成功，或者所有的选择都已用完时求值失败。
            - 这要求求值器保存另一个选择的引用，也就是维持有关选择的轨迹
      - amb分析器执行过程要求如下3个参数
         - 一个环境，执行过程的时候都需要的。
         - 一个成功继续
         - 一个失败继续
      - amb分析器要注意
         - 赋值会改变，如果要回到失败继续，那么相应的赋值也要恢复。
